```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("student_clustering.csv")
```

```python
data
```

```python
plt.scatter(data["cgpa"], data["iq"], label = "DataSet")
plt.xlabel("cgpa")
plt.ylabel("IQ")
plt.legend()
```

```python
# Elbow Methods
```

```python
from sklearn.cluster import KMeans
```

```python
check_wcss = KMeans(n_clusters = 1)

check_wcss.fit_predict(data)

check_wcss.inertia_
```

```python
check_wcss = KMeans(n_clusters = 2)

check_wcss.fit_predict(data)

check_wcss.inertia_
```

```python
check_wcss = KMeans(n_clusters = 3)

check_wcss.fit_predict(data)

check_wcss.inertia_
```

```python
check_wcss = KMeans(n_clusters = 4)

check_wcss.fit_predict(data)

check_wcss.inertia_
```

```python
check_wcss = KMeans(n_clusters = 5)

check_wcss.fit_predict(data)

check_wcss.inertia_
```

```python
check_wcss = KMeans(n_clusters = 6)

check_wcss.fit_predict(data)

check_wcss.inertia_
```

```python
wcss = []
```

```python
for i in range(1, 10):
    check_wcss = KMeans(n_clusters = i)
    check_wcss.fit_predict(data)
    wcss.append(check_wcss.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 10), wcss, label = "Elbow Curve", marker = ".", ms = 15, mec = "green", mfc = "yellow")
plt.xlabel("number of cluster")
plt.ylabel("WCSS / Inertia..")
plt.legend()

for i in range(1, 10):
    plt.annotate(xy = (i, wcss[i-1]), text = f"({i}, {str(wcss[i-1])[:5]})")
```
